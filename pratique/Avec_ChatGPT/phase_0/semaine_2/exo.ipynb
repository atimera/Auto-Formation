```python
import numpy as np

# une matrice X (3x2)
# afficher shape
# faire transpose
# faire multiplication

X = np.array([
    [1, 2],
    [2, 3],
    [3, 4]
])

Y = np.array([
    [1,2,3],
    [4,5,6]
])

print("X = \n", X)
print("Shape de X = ", X.shape)
print("Transposition de X = \n", X.T)
print("Shape de Y = ", Y.shape)
print("Multiplication matricielle X * Y = \n", np.dot(X, Y))
```
