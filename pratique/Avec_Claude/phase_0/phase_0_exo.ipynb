```python
import numpy as np

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print("Produit scalaire :", np.dot(v1, v2))

A = np.array([
    [1, 2], 
    [3, 4]
])
B = np.array([
    [5, 6], 
    [7, 8]
])

print("Produit matriciel :\n", np.matmul(A, B))
print("Produit matriciel 2 :\n", A * B)
print("Transposée de A :\n", A.T)
```

```python
import numpy as np

A = np.array([1,2])
B = np.array([3, 4])

print("A + B = ", A + B)
print("A - B = ", A - B)
print("2A - B = ", 2*A - B)
print("A . B = ", np.dot(A, B))
print("A * B = ", A * B)
print("A.T = ", A.T) 
```

```python
import numpy as np

A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

B = np.array([
    [1,2],
    [3,4],
    [5,6]
])

C = np.array([
    [1,3,5],
    [2,4,6]
])

print("A = \n", A)
print("B = \n", B)
print("C = \n", C)
print("------------")
print("A.T = \n", A.T)
print("2A = \n", 2*A)
print("A * B = \n", A * B)
#print("np.matmul(A,B) = \n", np.matmul(A,B))
#print("C * B = \n", C * B) -> Impossible apparemment
print("C * B = \n", np.matmul(C,B))
```

```python
import numpy as np

A = np.array([[1,2],[3,4],[5,6]])   # 3×2
B = np.array([[1,2],[3,4],[5,6]])   # 3×2
C = np.array([[1,3,5],[2,4,6]])     # 2×3

print("A * B (terme à terme) =\n", A * B)       # ✅ même dimension
print("A @ C (matriciel) =\n", A @ C)            # ✅ (3×2) @ (2×3) = (3×3)
print("C @ A (matriciel) =\n", C @ A)            # ✅ (2×3) @ (3×2) = (2×2)
print("C @ B (matriciel) =\n", C @ B)            # ✅ (2×3) @ (3×2) = (2×2)
```
